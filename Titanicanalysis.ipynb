```python
# Drop unnecessary columns
titanic_df_cleaned = titanic_df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
```

```python
titanic_df_cleaned
```

```python
# Rename columns for better readability
titanic_df_cleaned.rename(columns={
    'Survived': 'survived',
    'Pclass': 'class',
    'Sex': 'gender',
    'Age': 'age',
    'SibSp': 'siblings_spouses',
    'Parch': 'parents_children',
    'Fare': 'fare',
    'Embarked': 'embarked'
}, inplace=True)
```

```python
titanic_df_cleaned
```

```python
# Remove duplicate rows
titanic_df_cleaned.drop_duplicates(inplace=True)
```

```python
titanic_df_cleaned
```

```python
# Convert categorical variables to numerical values
titanic_df_cleaned['gender'] = titanic_df_cleaned['gender'].map({'male': 0, 'female': 1})
titanic_df_cleaned['embarked'] = titanic_df_cleaned['embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```

```python
titanic_df_cleaned
```

```python
# Fill NaN values in 'age' with the median age
titanic_df_cleaned['age'].fillna(titanic_df_cleaned['age'].median(), inplace=True)
```

```python
titanic_df_cleaned
```

```python
# Fill NaN values in 'embarked' with the most common embarkation point
titanic_df_cleaned['embarked'].fillna(titanic_df_cleaned['embarked'].mode()[0], inplace=True)
```

```python
titanic_df_cleaned
```

```python
# Verify cleaning steps
titanic_df_cleaned.info()
```

```python
# Descriptive statistics
titanic_df_cleaned.describe()
```

```python
# Count the number of unique values in each column
titanic_df_cleaned.nunique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df_cleaned['age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot of Gender
plt.figure(figsize=(10, 6))
sns.countplot(x='gender', data=titanic_df_cleaned)
plt.title('Count of Each Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Box plot of Age by Class
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='age', data=titanic_df_cleaned)
plt.title('Box Plot of Age by Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
# Heatmap of correlations
plt.figure(figsize=(10, 6))
sns.heatmap(titanic_df_cleaned.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Group by class and calculate mean
class_grouped = titanic_df_cleaned.groupby('class').mean()
class_grouped
```

```python
# Group by gender and calculate mean
gender_grouped = titanic_df_cleaned.groupby('gender').mean()
gender_grouped
```

```python
# Group by embarked and calculate mean
embarked_grouped = titanic_df_cleaned.groupby('embarked').mean()
embarked_grouped
```

```python
# Check for missing values
missing_values = titanic_df_cleaned.isna().sum()
missing_values
```

```python
# Save the cleaned dataset to a CSV file
titanic_df_cleaned.to_csv('cleaned_titanic_data.csv', index=False)
```

```python
# Survival rate by embarkation point
survival_rate_by_embarked = titanic_df_cleaned.groupby('embarked')['survived'].mean()
survival_rate_by_embarked
```
